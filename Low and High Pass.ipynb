```python
class UnivariateDataset(PastCovariatesTrainingDataset):
    def __init__(self, inlen, outlen, data, seasonality = 12, shift_time = False):
        super().__init__()
        
        
        self.inlen = inlen
        self.outlen = outlen
        self.seasonality = seasonality
        self.shift_time = shift_time
        self.data = data.ravel()
        self.dataset = self.__create_dataset(data, inlen, outlen)
        
        # length of the origiginal dataset
        self.orig_len = len(self.dataset)
        
    def __getitem__(self, idx: int):
        return self.dataset[idx]

    def __len__(self):
        return len(self.dataset)
    
    # returns the length of the original dataset without augmentations
    def original_len(self):
        return self.orig_len
    
    def __create_dataset(self, data, inlen, outlen):
        dataset = []

        for i in range(len(data)-(inlen+outlen)+1):
            if self.shift_time == True:
                x = np.expand_dims(data[i:i+inlen],1)
                y = np.expand_dims(data[i+(inlen-outlen):i+(2*inlen-outlen)],1)
            else:
                x = np.expand_dims(data[i:i+inlen],1)
                y = np.expand_dims(data[i+inlen:i+inlen+outlen],1)
                                        
            dataset.append((x,None,None,y))
        return dataset
```

### Original Signal

```python
size = 201

x = np.linspace(-np.pi*2, np.pi*2, size)
sin_x = np.sin(x)
plt.plot(x, sin_x)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```

### Noisy signal

```python
white_noise = np.random.normal(0,0.1,size = size)
noisy_sin_x = sin_x + white_noise 

plt.plot(x, noisy_sin_x)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```

### Low Pass - Smoothing/Denoising

```python
def low_pass(signal, kernel_size):
    kernel = np.ones((kernel_size))*(1/kernel_size)
    output =  np.convolve(signal,kernel,mode = 'same')
    return output  
    
```

#### Different kernel sizes

```python

smooth1 = low_pass(noisy_sin_x, kernel_size = 4)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth1)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("kernel_size = 4")
plt.show()

smooth2 = low_pass(noisy_sin_x, kernel_size = 10)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth2)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("kernel_size = 10")
plt.show()


smooth3 = low_pass(noisy_sin_x, kernel_size = 50)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth3)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("kernel_size = 50")
plt.show()
```

```python

smooth = low_pass(noisy_sin_x, kernel_size = 4)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("pass 1")
plt.show()

smooth = low_pass(smooth, kernel_size = 4)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("pass 2")
plt.show()

smooth = low_pass(smooth, kernel_size = 4)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("pass 3")
plt.show()

smooth = low_pass(smooth, kernel_size = 4)
plt.plot(x, noisy_sin_x)
plt.plot(x, smooth)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.title("pass 4")
plt.show()
```

### High pass - sobel filter

```python
#note: instead of [-1,1] we apply [1,-1] and instead o of [1,-2,1]we apply [-1,2,-1]
# which has an equivalent outcome but a different sign 
def high_pass(signal, d = 1, eps = 1):
    if d==1:
        kernel = np.zeros((eps+1))  
        kernel[0] = 1
        kernel[-1] = -1

    if d==2:
        kernel = kernel = np.zeros((2*eps+1))  
        kernel[0] = -1
        kernel[-1] = -1
        kernel[eps] = 2
            
    output =  np.convolve(signal,kernel,mode = 'same')
    return output  
```
